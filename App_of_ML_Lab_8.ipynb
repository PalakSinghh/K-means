# LAB 8 EM & K-MEANS
**Apply EM algorithm to cluster a set of data stored in a .CSV file. Use the same data set for clustering using k-Means algorithm. Compare the results of these two algorithms and comment on the quality of clustering.**

```python
import numpy as np
import pandas as pd
from sklearn.mixture import GaussianMixture
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Load the data from CSV file
data = pd.read_csv('/content/faithful.csv')
data = data.drop('Unnamed: 0', axis=1)

# Convert the data to numpy array
X = np.array(data)

# Normalize the data
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)

# Apply EM algorithm
em = GaussianMixture(n_components=2, max_iter=100)
em.fit(X)
em_labels = em.predict(X)

# Apply k-Means algorithm
kmeans = KMeans(n_clusters=2, max_iter=100)
kmeans.fit(X)
kmeans_labels = kmeans.labels_

# Plot the results
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
ax1.scatter(X[:, 0], X[:, 1])
ax1.set_title('Original Data')
ax2.scatter(X[:, 0], X[:, 1], c=em_labels)
ax2.set_title('EM Algorithm')
ax3.scatter(X[:, 0], X[:, 1], c=kmeans_labels)
ax3.set_title('K-Means Algorithm')
plt.show()


# Normalize the data
scaler = StandardScaler()
X_norm = scaler.fit_transform(X)

# Apply EM algorithm
em = GaussianMixture(n_components=2, max_iter=100)
em.fit(X_norm)
em_labels = em.predict(X_norm)

# Apply k-Means algorithm
kmeans = KMeans(n_clusters=2, max_iter=100)
kmeans.fit(X_norm)
kmeans_labels = kmeans.labels_

# Plot the results with normalized data
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 5))
ax1.scatter(X[:, 0], X[:, 1])
ax1.set_title('Original Data')
ax2.scatter(X[:, 0], X[:, 1], c=em_labels)
ax2.set_title('EM Algorithm')
ax3.scatter(X[:, 0], X[:, 1], c=kmeans_labels)
ax3.set_title('K-Means Algorithm')
plt.show()

```
